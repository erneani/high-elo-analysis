# Prevendo resultado de partidas High-Elo no LoL

## Contexto

Buscando auxiliar os jogadores a compreenderem, em tempo de partida, as suas chances de vitória, iremos analisar um [dataset](https://www.kaggle.com/bobbyscience/league-of-legends-diamond-ranked-games-10-min) com mais de 10 mil partidas de jogadores de Rank Diamante (um dos mais altos do jogo) e como seus comportamentos nos primeiros 10 minutos do jogo podem definir o sucesso de seu time (e, consequentemente, a valorização de suas ações).

## Mãos à obra

```python
# Antes de tudo, precisamos importar nossas bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Uma vez que importamos nossas bibliotecas, precisamos importar agora o nosso Dataset
# Utilizaremos a função de ler CSV do pandas para isso
df = pd.read_csv('high_diamond_ranked_10min.csv')
```

```python
# Após a leitura do dataset, checaremos algumas informações estatísticas básicas
# utilizando as funções describe() e info() (ambas do Pandas)
df.describe()
```

```python
df.info()
```

```python
# Observe que temos 39 colunas (18 do time azul, 18 do time vermelho, 
# 2 colunas que especificam o ID da partida e 1 coluna que especifica o resultado)
# Vamos ver as primeiras linhas do nosso DataFrame utilizando o comando head()
df.head()
```

```python
# É importante verificar, antes de tudo, se há dados em branco (ou nulos)
# em nosso DataFrame para poder, de maneira correta, extrair as informações
# necessárias para o nosso algoritmo
sns.heatmap(df.isnull())
```

```python
# Como não há informações em branco ou nulas, vamos checar as colunas
df.columns
```

```python
# Observemos a quantidade de vitórias do time azul
sns.countplot(df['blueWins'])
```

```python
# Como não há quase uma equivalência nas vitórias, não podemos considerar
# a cor do time como algo decisivo para o jogo (uma coisa óbvia estatisticamente comprovada).
# Sabendo disso, vamos checar, agora, a correlação entre a taxa de vitória (dos azuis)
# e as outras informações do dataset.
# Note que os valores foram ordenados do maior para o menor para facilitar a visualização
# de informações que realmente levam um jogo a ser ganho
df.corr()['blueWins'].sort_values(ascending=False)
```

```python
# Quando observamos as informações, podemos notar que existem colunas que influenciam
# muito pouco na vitória
# Essas colunas são o gameId e blueWardsPlaced
# Vamos remover o gameId
# PS: não removeremos o blueWardsPlaced por possuir uma coluna
# chamada redWardsPlaced que possui correlação significante (e ambas são opostas).
df.drop('gameId', axis=1, inplace=True)
```

```python
# Podemos observar outras relações interessantes em nosso dataset
# As colunas redDeaths e blueKills são equivalentes,
# assim como as colunas:
# - blueDeaths e redKills
# - blueTotalMinionsKilled e blueCSPerMin
# - blueTotalGold e blueGoldPerMin
# - redTotalMinionsKilled e redCSPerMin
# - redTotalGold e redGoldPerMin
# Sabendo disso, poderemos remover esses dados duplicados.
# Para tal, escolheremos somente um dos elementos equivalentes de cada uma
# das duplas listadas.

equivalent_removal_list = ['blueDeaths', 'blueTotalMinionsKilled', 'blueTotalGold',
                          'redDeaths', 'redTotalMinionsKilled', 'redTotalGold']

df.drop(equivalent_removal_list, axis=1, inplace=True)
```

```python
# Agora que removemos as colunas equivalentes ou que não são importantes
# para analisar nossa partida, vamos criar nosso modelo para o
# Aprendizado de Máquina!

# Antes de tudo, vamos separar nossos dados em dados de teste e dados de treino,
# permitindo que façamos o processo de *Aprendizagem Supervisionada*.
from sklearn.model_selection import train_test_split
```

```python
# Visando separar nosso dados para testar e treinar nosso modelo, 
# criaremos as variáveis X e y, onde X é o conjunto de colunas que analisaremos
# e y é a coluna de resultados que desejamos alcançar.
X = df.drop('blueWins', axis=1) # Passaremos todas as informações, com exceção de blueWins, para o *X*
y = df['blueWins'] # Passaremos somente blueWins para o *y*, já que é a coluna que queremos prever.
```

```python
# A função train_test_split separa, de maneira automática, uma parcela dos dados
# para podermos treinar nosso modelo e outra parcela para podermos testar
# a efetividade do nosso modelo.
# Para isso, precisamos passar nosso dataset (somente com as colunas necessárias), o tamanho da parcela a ser
# separada para testarmos nosso modelo.
# A função retorna, então, 4 informações:
# - *X_train*, que são nossos dados de treino.
# - *X_test*, nossos dados para teste.
# - *y_train*, a parcela de resultados usada para treino.
# - *y_test*, a parcela de resultados utilizada para testar a precisão do nosso modelo.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
# Observe que, de fato, temos um conjunto de dados separado, agora
print('---------Dados de treino--------- \n\n', X_train)
print('---------Dados de teste---------\n\n', X_test)
```

```python
# Agora que já separamos nossos dados, vamos treinar nosso modelo
# Importaremos, do Scikit-learn, um modelo de Regressão Logística
from sklearn.linear_model import LogisticRegression
```

```python
# O motivo de utilizarmos esse tipo de Regressão é o fato
# de que queremos adivinhar um resultado que é binário (ou seja,
# assume somente dois valores), o que é exatamente o resultado
# de uma Regressão Logística (mais detalhes **aqui**).
```

```python
# Agora que importamos a Regressão Logística, criemos nosso modelo.
logmodel = LogisticRegression()
```

```python
# Uma vez que temos uma instância do nosso modelo, treinemos ela
# para poder testar se, de fato, nossa máquina aprendeu algo.
# Lembre-se de treiná-la com os **dados de treinamento**!
# ps: Ao treinar o modelo, pode surgir um aviso em seu notebook. Uma vez que
# o conteúdo deste aviso vai além do escopo desse artigo, ignore-o.
logmodel.fit(X_train, y_train)
```

```python
# Agora, tentaremos prever o resultado do nosso modelo utilizando
# os dados de teste (estaremos, efetivamente, testando o desempenho
# do nosso modelo).
predicted_data = logmodel.predict(X_test)
```

```python
# Após executar a função acima, temos os predições do nosso modelo
# para os dados de teste.
predicted_data
```

```python
# Poderemos, então, comparar o que nosso modelo previu com
# o que, de fato, ocorreu (lembram-se da nossa variável *y_test*?)
y_test
```

```python
# Para comparar o desempenho do nosso modelo, podemos utilizar
# ferramentas de métricas que já vem inclusas no próprio Scikit-learn
# e criar um gráfico para tornar as comparações mais simples.
```

```python
# Importaremos duas métricas estatísticas distintas:
# Matriz de confusão e o Relatório de Classificação.
from sklearn.metrics import confusion_matrix, classification_report
```

```python
# A Matriz de Confusão nos mostra a quantidade de Falsos Positivos, 
# Falsos Negativos, Positivos Verdadeiros e Negativos Verdadeiros
# no resultado de nosso dados ao comparar os dados previstos
# e os dados reais de teste.
# Mais detalhes sobre a Matriz de Confusão podem ser encontrados **aqui**.
print(confusion_matrix(predicted_data, y_test))
```

```python
# O Relatório de Classificação é um relatório um pouco mais
# complexo que mostra o desempenho do nosso modelo
# baseado em 3 metas estatísticas: a precisão, o recall e
# a pontuação F1.
print(classification_report(predicted_data, y_test))
```

```python
# Como podemos notar, nosso modelo possui cerca de
# 72% de precisão. Nada mal, uma vez que não houve
# uma exploração intensa acerca dos dados que
# realmente são vitais para as vitórias.
```
